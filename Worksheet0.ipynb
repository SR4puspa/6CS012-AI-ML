### 4.1 Exercise on Functions
#### Task 1: Unit Conversion Program

```python
def convert_length(value, direction):
    """
    Converts between meters and feet.

    Parameters:
        value (float): The numerical value to convert.
        direction (str): 'mtf' for meters to feet, 'ftm' for feet to meters.

    Returns:
        float: The converted value.
    """
    if direction == 'mtf':
        return value * 3.28084
    elif direction == 'ftm':
        return value / 3.28084
    else:
        raise ValueError("Invalid direction. Use 'mtf' or 'ftm'.")


def convert_weight(value, direction):
    """
    Converts between kilograms and pounds.

    Parameters:
        value (float): The numerical value to convert.
        direction (str): 'ktl' for kg to lbs, 'ltk' for lbs to kg.

    Returns:
        float: The converted value.
    """
    if direction == 'ktl':
        return value * 2.20462
    elif direction == 'ltk':
        return value / 2.20462
    else:
        raise ValueError("Invalid direction. Use 'ktl' or 'ltk'.")


def convert_volume(value, direction):
    """
    Converts between liters and gallons.

    Parameters:
        value (float): The numerical value to convert.
        direction (str): 'ltg' for liters to gallons, 'gtl' for gallons to liters.

    Returns:
        float: The converted value.
    """
    if direction == 'ltg':
        return value * 0.264172
    elif direction == 'gtl':
        return value / 0.264172
    else:
        raise ValueError("Invalid direction. Use 'ltg' or 'gtl'.")


def unit_converter():
    """
    Main function to drive the unit conversion program.
    Prompts the user to select a conversion type, input a value,
    and displays the converted result. Handles invalid input gracefully.

    Returns:
        float: The converted result, or None if input was invalid.
    """
    print("Unit Converter")
    print("Choose a conversion type:")
    print("1. Length  (meters <-> feet)")
    print("2. Weight  (kg <-> lbs)")
    print("3. Volume  (liters <-> gallons)")

    try:
        category = input("Enter category (1/2/3): ").strip()

        if category == '1':
            print("  a) Meters to Feet")
            print("  b) Feet to Meters")
            sub = input("Enter option (a/b): ").strip().lower()
            direction = 'mtf' if sub == 'a' else 'ftm'
            unit_in = 'meters' if sub == 'a' else 'feet'
            unit_out = 'feet' if sub == 'a' else 'meters'
            value = float(input(f"Enter value in {unit_in}: "))
            result = convert_length(value, direction)

        elif category == '2':
            print("  a) Kilograms to Pounds")
            print("  b) Pounds to Kilograms")
            sub = input("Enter option (a/b): ").strip().lower()
            direction = 'ktl' if sub == 'a' else 'ltk'
            unit_in = 'kg' if sub == 'a' else 'lbs'
            unit_out = 'lbs' if sub == 'a' else 'kg'
            value = float(input(f"Enter value in {unit_in}: "))
            result = convert_weight(value, direction)

        elif category == '3':
            print("  a) Liters to Gallons")
            print("  b) Gallons to Liters")
            sub = input("Enter option (a/b): ").strip().lower()
            direction = 'ltg' if sub == 'a' else 'gtl'
            unit_in = 'liters' if sub == 'a' else 'gallons'
            unit_out = 'gallons' if sub == 'a' else 'liters'
            value = float(input(f"Enter value in {unit_in}: "))
            result = convert_volume(value, direction)

        else:
            print("Invalid category selected.")
            return None

        print(f"\nResult: {value} {unit_in} = {result:.4f} {unit_out}")
        return result

    except ValueError as e:
        print(f"Error: Invalid input - {e}")
        return None



unit_converter()
```

#### Task 2: Mathematical Operations on a List of Numbers

```python
def compute_sum(numbers):
    """
    Computes the sum of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The sum of all elements in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("List is empty.")
    total = 0
    for n in numbers:
        total += n
    return total


def compute_average(numbers):
    """
    Computes the average (mean) of a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The average of all elements in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("List is empty.")
    return compute_sum(numbers) / len(numbers)


def compute_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The maximum value in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("List is empty.")
    max_val = numbers[0]
    for n in numbers:
        if n > max_val:
            max_val = n
    return max_val


def compute_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
        numbers (list): A list of numeric values.

    Returns:
        float: The minimum value in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("List is empty.")
    min_val = numbers[0]
    for n in numbers:
        if n < min_val:
            min_val = n
    return min_val


def math_operations_program():
    """
    Interactive program that performs mathematical operations on user-supplied numbers.
    Supports sum, average, maximum, and minimum. Handles invalid input and empty lists.

    Returns:
        float or None: The result of the selected operation, or None on error.
    """
    print("Math Operations Program")
    print("=" * 30)
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = input("Enter choice (1/2/3/4): ").strip()
        raw = input("Enter numbers separated by spaces: ")
        numbers = [float(x) for x in raw.strip().split()]

        ops = {'1': compute_sum, '2': compute_average,
               '3': compute_maximum, '4': compute_minimum}
        op_names = {'1': 'Sum', '2': 'Average', '3': 'Maximum', '4': 'Minimum'}

        if choice not in ops:
            print("Invalid choice.")
            return None

        result = ops[choice](numbers)
        print(f"\n{op_names[choice]}: {result}")
        return result

    except ValueError as e:
        print(f"Error: {e}")
        return None


# Empty list error handling
try:
    compute_sum([])
except ValueError as e:
    print(f"\nEmpty list error caught: {e}")


math_operations_program()
```

### 4.2 Exercise on List Manipulation

```python
# 1. Extract Every Other Element
def extract_every_other(lst):
    """Returns every other element starting from the first."""
    return lst[::2]

print("1. Extract every other:", extract_every_other([1, 2, 3, 4, 5, 6]))


# 2. Slice a Sublist
def get_sublist(lst, start, end):
    """Returns a sublist from index start to end (inclusive)."""
    return lst[start:end + 1]

print("2. Sublist [2:4]       :", get_sublist([1, 2, 3, 4, 5, 6], 2, 4))


# 3. Reverse a List Using Slicing
def reverse_list(lst):
    """Returns a reversed copy of the list using slicing."""
    return lst[::-1]

print("3. Reversed list       :", reverse_list([1, 2, 3, 4, 5]))


# 4. Remove the First and Last Elements
def remove_first_last(lst):
    """Returns the list without its first and last elements."""
    return lst[1:-1]

print("4. Remove first & last :", remove_first_last([1, 2, 3, 4, 5]))


# 5. Get the First n Elements
def get_first_n(lst, n):
    """Returns the first n elements of the list."""
    return lst[:n]

print("5. First 3 elements    :", get_first_n([1, 2, 3, 4, 5], 3))


# 6. Extract Elements from the End
def get_last_n(lst, n):
    """Returns the last n elements of the list."""
    return lst[-n:]

print("6. Last 2 elements     :", get_last_n([1, 2, 3, 4, 5], 2))


# 7. Extract Elements in Reverse Order (every second, starting from second-to-last)
def reverse_skip(lst):
    """Returns every second element starting from the second-to-last, moving backward."""
    return lst[-2::-2]

print("7. Reverse skip        :", reverse_skip([1, 2, 3, 4, 5, 6]))
```

### 4.3 Exercise on Nested List

```python
# 1. Flatten a Nested List
def flatten(lst):
    """Flattens a one-level nested list into a single list."""
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result

print("1. Flatten            :", flatten([[1, 2], [3, 4], [5]]))


# 2. Access Nested Element by Indices
def access_nested_element(lst, indices):
    """Extracts an element from a nested list given a list of indices."""
    element = lst
    for idx in indices:
        element = element[idx]
    return element

nested = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("2. Access [1][2]      :", access_nested_element(nested, [1, 2]))


# 3. Sum of All Elements in a Nested List
def sum_nested(lst):
    """Recursively sums all numeric elements in a nested list of arbitrary depth."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print("3. Sum nested         :", sum_nested([[1, 2], [3, [4, 5]], 6]))


# 4. Remove Specific Element from Nested List
def remove_element(lst, elem):
    """Removes all occurrences of elem from a one-level nested list."""
    return [[x for x in sublist if x != elem] for sublist in lst]

print("4. Remove 2           :", remove_element([[1, 2], [3, 2], [4, 5]], 2))


# 5. Find Maximum Element in a Nested List
def find_max(lst):
    """Finds the maximum element in a nested list of arbitrary depth."""
    max_val = None
    for item in lst:
        if isinstance(item, list):
            candidate = find_max(item)
        else:
            candidate = item
        if max_val is None or candidate > max_val:
            max_val = candidate
    return max_val

print("5. Max nested         :", find_max([[1, 2], [3, [4, 5]], 6]))


# 6. Count Occurrences in a Nested List
def count_occurrences(lst, elem):
    """Counts how many times elem appears in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

print("6. Count 2 in list    :", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))


# 7. Deep Flatten (any depth)
def deep_flatten(lst):
    """Recursively flattens a list of any nesting depth into a single list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print("7. Deep flatten       :", deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))


# 8. Nested List Average
def average_nested(lst):
    """Calculates the average of all elements in a nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print("8. Average nested     :", average_nested([[1, 2], [3, 4], [5, 6]]))
```

### 10.2 Experiment: How Fast is NumPy?

```python
SIZE = 1_000_000

# Pre-generate data
list_a = [float(i) for i in range(SIZE)]
list_b = [float(i) * 2 for i in range(SIZE)]
np_a = np.array(list_a)
np_b = np.array(list_b)

print("="*60)
print("EXPERIMENT: NumPy vs Python Lists")
print(f"Array size: {SIZE:,}")
print("="*60)

# 1. Element-wise Addition
# Python list
t0 = time.time()
result_list = [list_a[i] + list_b[i] for i in range(SIZE)]
t_list_add = time.time() - t0

# NumPy
t0 = time.time()
result_np = np_a + np_b
t_np_add = time.time() - t0

print(f"\n1. Element-wise Addition")
print(f"   Python List : {t_list_add:.6f} s")
print(f"   NumPy       : {t_np_add:.6f} s")
print(f"   Speedup     : {t_list_add/t_np_add:.1f}x")

# 2. Element-wise Multiplication
t0 = time.time()
result_list = [list_a[i] * list_b[i] for i in range(SIZE)]
t_list_mul = time.time() - t0

t0 = time.time()
result_np = np_a * np_b
t_np_mul = time.time() - t0

print(f"\n2. Element-wise Multiplication")
print(f"   Python List : {t_list_mul:.6f} s")
print(f"   NumPy       : {t_np_mul:.6f} s")
print(f"   Speedup     : {t_list_mul/t_np_mul:.1f}x")

# 3. Dot Product
t0 = time.time()
dot_list = sum(list_a[i] * list_b[i] for i in range(SIZE))
t_list_dot = time.time() - t0

t0 = time.time()
dot_np = np.dot(np_a, np_b)
t_np_dot = time.time() - t0

print(f"\n3. Dot Product")
print(f"   Python List : {t_list_dot:.6f} s")
print(f"   NumPy       : {t_np_dot:.6f} s")
print(f"   Speedup     : {t_list_dot/t_np_dot:.1f}x")

# 4. Matrix Multiplication (1000 x 1000)
MAT_SIZE = 1000
mat_a_list = [[float(i * MAT_SIZE + j) for j in range(MAT_SIZE)] for i in range(MAT_SIZE)]
mat_b_list = [[float(i * MAT_SIZE + j) for j in range(MAT_SIZE)] for i in range(MAT_SIZE)]
mat_a_np = np.array(mat_a_list)
mat_b_np = np.array(mat_b_list)

# Python list matrix multiplication
t0 = time.time()
result_mat_list = [[sum(mat_a_list[i][k] * mat_b_list[k][j]
                        for k in range(MAT_SIZE))
                    for j in range(MAT_SIZE)]
                   for i in range(MAT_SIZE)]
t_list_mat = time.time() - t0

# NumPy
t0 = time.time()
result_mat_np = np.dot(mat_a_np, mat_b_np)
t_np_mat = time.time() - t0

print(f"\n4. Matrix Multiplication ({MAT_SIZE}x{MAT_SIZE})")
print(f"   Python List : {t_list_mat:.4f} s")
print(f"   NumPy       : {t_np_mat:.6f} s")
print(f"   Speedup     : {t_list_mat/t_np_mat:.1f}x")

print("\n" + "="*60)
print("Conclusion: NumPy is dramatically faster due to vectorized")
print("C-level operations, avoiding Python's interpreter overhead.")
print("="*60)
```
